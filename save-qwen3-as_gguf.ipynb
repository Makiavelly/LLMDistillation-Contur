```python
import subprocess

subprocess.run([
    "pip", "install", "-q",
    "unsloth==2026.4.6",
    "unsloth-zoo==2026.4.8",
    "transformers==5.5.0",
    "trl==0.24.0",
    "peft==0.19.1",
    "bitsandbytes==0.49.2",
    "accelerate==1.13.0",
    "tokenizers==0.22.2",
    "datasets==4.3.0",
])
```

```python
from unsloth import FastModel

model, tokenizer = FastModel.from_pretrained(
    model_name = "unsloth/Qwen3-4B",
    load_in_4bit = False,
)
```

```python
from peft import PeftModel

lora_path = "/kaggle/input/datasets/mishashalagin/my-checkpoint/checkpoint-200"
lora_model = PeftModel.from_pretrained(model, lora_path)
```

```python
merdeg_model = lora_model.merge_and_unload()
```

```python
merdeg_model.save_pretrained_gguf(
    "kaggle/working/qwen3-countdown-gguf",
    tokenizer,
    quantization_method="q4_k_m",
)
```
